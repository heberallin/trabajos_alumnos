```python
## se crea el contenedor subtotal que seria la suma de todos los montos de las distintas cuentas por pagar
## ademas se crean los contenedores ingreso y saldo que sera la diferencia entre ingreso y subtotal
```

```python
items= df["tipo de cuenta"]
montos=df["monto"]
plt.subplot(111)
plt.pie(montos, labels=items,autopct="%1.1F%%")
plt.axis("equal")
plt.title("grafico de cuentas por pagar")
plt.subplots_adjust(left=0, bottom=0,right=2,top=2.2,wspace=0.6,hspace=0.6)
plt.show
```
