```python
# importamos nuestro archivo de coeficiente de cultivo el cual nos mostrara el indice , 
# el cual lo utilizaremos para para realizar la busqueda de coeficiente de cultivo
# asociado a cada etapa del desarrollo 

cultivo = pd.read_excel("C:/Users/arnaldo/Desktop/pyton/coeficientecultivo.xlsx")
cultivo.head(24)






```

```python
# se relaciona el indice que proporciona pandas asociandolo
#al cultivo como un entero, se ingresa el desarrollo del cultivo 
#como un string, el resultado nos indica el coeficiente de cultivo "Kc"


a = int(input("ingrese indice cultivo :")) # se ingresa el indice asociado al cultivo
b = str(input("Ingrese Desarrollo :"))



Kcultivo = (cultivo.loc[a].at[b])
print("coeficiente de cultivo es ", (round(Kcultivo,2)))



```

```python
# importamos nuestro archivo de coeficiente de cultivo de frutas el cual nos mostrara el indice , 
# el cual lo utilizaremos para para realizar la busqueda de coeficiente de cultivo
# asociado a cada mes de cosecha
frutas = pd.read_excel("C:/Users/arnaldo/Desktop/pyton/coeficientefrutas.xlsx")
frutas.head(24)

```

```python
# se relaciona el indice que proporciona pandas asociandolo
#ala fruta como un entero, se ingresa la epoca para cosechar (meses) 
#como un string, el resultado nos indica el coeficiente de cultivo "Kc"
#asociado al cultivo de frutas
c = int(input("ingrese Fruta :"))
d = str(input("ingrese Mes : "))


Kcfruta = float(frutas.loc[c].at[d])

if Kcfruta <= 0: 
 print("Epoca de año no cultivable", Kcfruta)

else:
 print("coeficiente de cultivo para la fruta seleccionada es ", Kcfruta)
```

```python
# importamos el archivo historico de evapotraspiracion referencial 
# contemplando mensualmente la evapotraspiracion en el rango de 
# años del 2015 al año 2018
eva = pd.read_excel("C:/Users/arnaldo/Desktop/pyton/agua.xlsx")
eva.head(400)







```

```python
# el indice que asocia pandas a cada fila lo utilizaremos para asociarlo a cada mes 
# enero seria 0 .... ingresandolo como un entero, aun que se puede trabajar como 
# dato flotante 
# la variable est, se usara un string para seleccionar la estacion mas cercana a la propiedad
# en la aplicacion real se debera mostrar una referencia de la ubicacion 
# de la estacion meterelogica


x = int(input("ingrese mes:"))
est = str(input("estacion:"))



Etr = int(eva.loc[x].at[est])

print("evapotraspiracion referencial para el mes elegido es ", (round(Etr,2)), "mm/mes")





```

```python
# Calculo Evapotraspiracion real en milimetros dia [mm d-1]
# ETreal = Etr * Kc
# Etr = evapotraspiracion referencial
# Kc = Coeficiente cultivo

Etreal = Etr * Kcultivo  # Evapotraspiracion real para cultivo

EtrealFr = Etr * (Kcfruta)  # evapotraspiracion real para cultivo
EtrealFr = (EtrealFr)

Etrealdiarioctvo = (Etreal/30)
EtrealdiarioFr = (EtrealFr/30)



print("Evapotraspiracion real para el cultivo seleccionado", (round(Etreal,2)), "mm/mes")# dado que se utiliza un promedio mensual 
# la Etreal es en milimetros mensual 
print("Evapotraspiracion real para la fruta seleccionada",(round(EtrealFr,2)), "mm/mes")
print("Evapotraspiracion real para el cultivo seleccionado", (round(Etrealdiarioctvo,2)), "mm/dia")
print("Evapotraspiracion real para la fruta seleccionada",(round(EtrealdiarioFr,2)), "mm/dia")# redondeando 
```
