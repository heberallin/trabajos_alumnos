```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_excel("/Users/andres/Desktop/UAC CUARTO SEMESTRE /base datos python .xlsx")
df.head(12)


df2 = df.drop(["MES", "Y","C", "T", "LTI_C", "RWC_C", "MTC_C", "FAC_C"], axis = 1)

df2.head()

df_y = df.drop(["MES", "Y","C", "T", "LTI_C", "RWC_C", "MTC_C", "FAC_C"], axis = 1)

df_c = df.drop(["MES", "Y","C", "T", "LTI_Y", "RWC_Y", "MTC_Y", "FAC_Y"], axis = 1)


df_y["eventos"] = df_y.sum(axis=1)
df_c["eventos"] = df_c.sum(axis=1)
df_y.head(12)


plt.plot(df["MES"], df_y["eventos"],marker="o")
plt.xticks(rotation=90)
plt.ylabel("Nº de eventos")
plt.xlabel("Mes del año 2018")
plt.title("Eventos de Yamana en el año 2018")
plt.show()
print("El numero de eventos el año 2018 por Yamana fueron: ", int(df_y["eventos"].sum()))


plt.plot(df["MES"], df_c["eventos"],marker="o")
plt.xticks(rotation=90)
plt.ylabel("Nº de eventos")
plt.xlabel("Mes del año 2018")
plt.title("Eventos de Contratistas en el año 2018")
plt.show()
print("El numero de eventos el año 2018 por Contratistas fueron: ", int(df_c["eventos"].sum()))

print("El numero de eventos el año 2018 por Yamana + Contratistas fueron: ", int(df_y["eventos"].sum())+ int(df_c["eventos"].sum()))
```

**Reporte grafico de horas trabajadas (HH)**

```python
plt.plot(df["MES"], df["Y"],"ro--",linestyle="dotted")
plt.xticks(rotation=90)
plt.xlabel("Mes año 2018")
plt.ylabel("Nº horas trabajadas")
plt.title("Horas trabajadas por Yamana en el año 2018")
plt.show()
print ("El promedio de horas trabajadas por yamana elaño 2018 es de:", xy)
print ("El total de horas trabajadas por Yamana el año 2018 es de: " , total_y)

plt.plot(df["MES"], df["C"],"bo--",linestyle="dotted")
plt.xticks(rotation=90)
plt.xlabel("Mes año 2018")
plt.ylabel("Nº horas trabajadas")
plt.title("Horas trabajadas por Colaboradores en el año 2018")
plt.show()
print ("El promedio de horas trabajadas por empresas colaboradoras el año 2018 es de:", xc)
print ("El total de horas trabajadas por Colaboradores el año 2018 es de: " , total_c)

plt.plot(df["MES"], df["T"],"go--",linestyle="dotted")
plt.xticks(rotation=90)
plt.xlabel("Mes año 2018")
plt.ylabel("Nº horas trabajadas")
plt.title("Horas trabajadas por Yamana + Colaboradores en el año 2018")
plt.show()
print ("El promedio de horas trabajadas por Yamana + Colaboradores el año 2018 es de:", xt)
print ("El total de horas trabajadas por Yamana + Colaboradores  el año 2018 es de: " , total_y + total_c)

```
